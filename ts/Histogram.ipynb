# Codes below are implemented in R in the video lecture

```r
data <- c(91, 49, 76, 112, 97, 42, 70, 100, 8, 112, 95, 90, 78, 62, 56, 94, 65, 58, 109, 70, 109, 91, 71, 76, 68, 62, 134, 57, 83, 66)
```

```r
hist(data)
```

```r
hist(data, xlab = "My data points")
```

```r
hist(data, xlab = "My data points", main = "Histogram of my data")
```

```r
hist(data, xlab = "My data points", main = "Histogram of my data", freq = F)
```

```r
hist(data, xlab = "My data points", main = "Histogram of my data", freq = F, col = "green")
```

```r
hist(data, xlab = "My data points", main = "Histogram of my data", freq = F, col = "green")
lines(density(data))
```

```r
hist(data, xlab = "My data points", main = "Histogram of my data", freq = F, col = "green")
lines(density(data), col = "red", lwd = 5)
```

```r
hist(data, xlab = "My data points", main = "Histogram of my data", freq = F, col = "green", breaks = 10)
```

```r
hist(data, xlab = "My data points", main = "Histogram of my data", freq = F, col = "green", breaks = 10)
lines(density(data), col = "red", lwd = 5)
```
